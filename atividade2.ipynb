2.1. Você trabalha em uma transportadora urbana que opera 3 tipos de veículos urbanos de carga
(VUC):
O modelo 1, da marca JAC, tem 38 m3 de volume útil e consumo de 9 km/l;
O modelo 2, da marca KIA, tem 44 m3 de volume útil e consumo de 6 km/l;
O modelo 3, da marca Mercedes-Benz, tem 42 m3 de volume útil e consumo de 7 km/l.


A empresa realiza somente fretes cheios, ou seja, com o VUC com carga próxima a máxima
permitida.
A transportadora oferece caixas de transporte com as seguintes dimensões:

*   Caixa modelo A com 0,5 m3 de capacidade;
*   Caixa modelo B com 1,0 m3 de capacidade;
*   Caixa modelo C com 2,0 m3.

Elabore um algoritmo em Python que receba (entrada de dados) a quantidade de cada tipo de caixa a ser transportada e a distância a ser percorrida em quilômetros. O algoritmo deve determinar (processamento de dados) o volume total a ser transportado e escolher o caminhão (if...elif...else) que pode fazer a entrega gastando a menor quantidade de combustível e deveapresentar (saída) o modelo de caminhão escolhido, o volume total a ser transportado e o gasto esperado de combustível.

```python
# Entrada de dados das caixas
qtdd_carga1 = float(input('Digite a quantidade de carga da caixa modelo A: ')) * 0.5
qtdd_carga2 = float(input('Digite a quantidade de carga da caixa modelo B: ')) * 1.0
qtdd_carga3 = float(input('Digite a quantidade de carga da caixa modelo C: ')) * 2.0


# Mostrar a quantidade em m3 de capacidade de cada carga
print(f'O total em m³ da caixa A é: {qtdd_carga1}, da caixa B é: {qtdd_carga2}, da caixa C é: {qtdd_carga3}')

total_caixas = qtdd_carga1 + qtdd_carga2 + qtdd_carga3
print(f'A quantidade em m3 de todas as caixas é: {total_caixas}')

# Entrada da distância
distancia = float(input('Digite a distância a ser percorrida em quilômetros: '))

# Saída do consumo de combustível formatada
print(f'O consumo total em litros do caminhão A: {consumo_veiculo1:.2f}, caminhão B: {consumo_veiculo2:.2f}, caminhão C: {consumo_veiculo3:.2f}')


# Processamento/saída

veiculo1 = 'Deve ser escolher o caminhão do modelo A da marca JAC '
veiculo2 = 'Deve ser escolher o caminhão do modelo B da marca KIA'
veiculo3 = 'Deve ser escolher o caminhão do modelo C da marca Mercedes-Benz'

consumo_veiculo1 = distancia / 9
consumo_veiculo2 = distancia / 6
consumo_veiculo3 = distancia / 7

# Verifica se a quantidade de m3 é menor ou igual a 38
if total_caixas <= 38:
  #Verifica se o consumo 1 é menor que o consumo 2 e 3
  if consumo_veiculo1 < consumo_veiculo2 and consumo_veiculo1 < consumo_veiculo3:
    print(veiculo1)
    #Verifica se o consumo 2 é a menor que o consumo 1 e o 3
  elif consumo_veiculo2 > consumo_veiculo1 and consumo_veiculo2 > consumo_veiculo3:
    print(veiculo2)
  else:
    print(veiculo3)
# Verifica se a quantidade de m3 é maior que 38 e menor que 42
elif total_caixas > 38 and total_caixas <= 42:
  if consumo_veiculo2 > consumo_veiculo1 and consumo_veiculo2 > consumo_veiculo3:
    print(veiculo2)
  else:
    print(veiculo3)
#Verifica se a quantidade de m3 é maior que 42 e menor que 44
elif total_caixas > 42 and total_caixas <= 44:
    print(veiculo3)
#Caso seja maior que 44 então a menssagem abaixo é exibida.
else:
    print('Nenhum caminhão serve')
```
